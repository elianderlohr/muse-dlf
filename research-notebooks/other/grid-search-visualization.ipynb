```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('grid_search_metrics.csv')
```

```python
# Extract relevant columns
df_filtered = df[['combination', 'epoch', 'f1_combined_micro', 'f1_combined_macro']]

# Group by combination
grouped = df_filtered.groupby('combination')

# Plotting
plt.figure(figsize=(15, 8))

for name, group in grouped:
    plt.plot(group['epoch'], group['f1_combined_micro'], label=f'{name} - Micro')
    plt.plot(group['epoch'], group['f1_combined_macro'], label=f'{name} - Macro', linestyle='--')

plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.title('F1 Combined Score Development Over Epochs by Combination')
plt.legend()
plt.show()
```

```python
# get the best combination based on the f1_combined_micro and f1_combined_macro score
best_micro = df_filtered.loc[df_filtered['f1_combined_micro'].idxmax()]
best_macro = df_filtered.loc[df_filtered['f1_combined_macro'].idxmax()]

print(f'Best combination based on f1_combined_micro: {best_micro["combination"]}')
print(f'Best combination based on f1_combined_macro: {best_macro["combination"]}')
```

```python
# get the best parameter from the best combination
best_param = df.loc[df['f1_combined_micro'].idxmax()]
print(f'Best parameter: {best_param}')
```
